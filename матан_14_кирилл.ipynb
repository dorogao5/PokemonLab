1. Программа которая по заданной формуле членов последовательности рисует
набор из N точек вида $(k, x_{k})$, где $k = [1, N] ∩ \mathbb{N}$.


```python
# Импорт необходимых библиотек
import numpy as np
from matplotlib import pyplot as plt

# Определяем функцию f для последовательности по новой формуле
def f(x):
    # Calculate the new formula
    term1 = np.arcsin((1 + (-1) ** x) / 2)  # Compute arcsin term
    term2 = (6 * x - 5) / (2 * x + 4)  # Fractional term
    return term1 * term2  # Multiply the terms

# Генерируем массив натуральных чисел от 1 до 100
x = np.array(list(range(1, 101)))
y = f(x)

# Создаем график
plt.figure(figsize=(12, 6), facecolor='#092b09')  # Темный зеленый фон

ax = plt.gca()  # Получаем текущие оси
ax.set_facecolor('#041804')  # Очень темный зеленый фон для осей

plt.scatter(x, y, color='#207a20', label='Значения последовательности', edgecolor='black', alpha=0.7)  # Мягкий темно-зеленый для точек
plt.plot(x, y, color='#207a20', label='Значения последовательности', alpha=0.7)  # Соединение точек линией мягким зеленым
plt.scatter(25, f(25), color='cyan', s=52, edgecolor='white', label='$x_{n}$ точка (для примера)')  # Подсветка точки с красным цветом

# Горизонтальная линия для справки
plt.axhline(y=0, color='darkgreen', linestyle='--', linewidth=0.8, alpha=0.6)  # Темный зеленый для горизонтальной линии

# Заголовки и метки осей
plt.title('Визуализация последовательности $x_{n}=\\arcsin\\frac{1 + (-1)^n}{2} \\cdot \\frac{6n - 5}{2n + 4}$', fontsize=14, color='#90ee90')
plt.xlabel('Значения $n \in \mathbb{N}$', fontsize=15, color='#90ee90')
plt.ylabel('Значения $x_{n}$', fontsize=12, color='#90ee90')

# Добавление сетки
plt.grid(visible=True, which='both', color='darkgreen', linestyle='--', linewidth=0.7, alpha=0.5)

# Показ значений на осях
plt.xticks(np.arange(0, 110, 10), color='#90ee90')  # Шаг значений по оси x в светло-зеленом
plt.yticks(np.linspace(min(y) - 2, max(y) + 3, 5), color='#90ee90')  # Шаг значений по оси y в светло-зеленом

# Легенда с изменением цвета текста
legend = plt.legend(facecolor='#092b09')
for text in legend.get_texts():
    text.set_color('#90ee90')  # Установка цвета текста легенды на светло-зеленый

# Показ графика
plt.show()
```

4-5. По данному $\varepsilon>0$ определим номер $n_{0}$, начиная с которого, элементы подпоследовательности будут попадать в $\varepsilon \text{-окрестность}$ предела. Отедельно выделим те точки, которые попадают в данную $\varepsilon \text{-окрестность}$ и отметьте горизонтальной линией значение предела этой подпоследовательности.


1) Пусть $ɛ = 0.1$:

```python
# Импорт необходимых библиотек
import numpy as np
from matplotlib import pyplot as plt
num_X1 = 81
num_X2 = 151
# Определяем функцию f для последовательности
def f(x):
    # Calculate the new formula
    term1 = np.arcsin((1 + (-1) ** x) / 2)  # Compute arcsin term
    term2 = (6 * x - 5) / (2 * x + 4)  # Fractional term
    return term1 * term2  # Multiply the terms

# Генерируем массив натуральных чисел от 1 до 100
x = np.array(list(range(num_X1, num_X2)))
y = f(x)

# Определяем последовательность для четных индексов 2k
x_even = np.array([2 * k for k in range(num_X1//2, num_X2//2)])  # Первые 50 четных индексов
y_even = f(x_even)

# Устанавливаем значение eps и предел подпоследовательности
epsilon = 0.1
limit = (3 * np.pi) / 2

# Находим n0 для подпоследовательности
indices_within_epsilon = np.where(np.abs(y_even - limit) < epsilon)[0]
n0_index = indices_within_epsilon[0] if indices_within_epsilon.size > 0 else None

# Создаем график
plt.figure(figsize=(12, 6), facecolor='#092b09')  # Темный зеленый фон
ax = plt.gca()  # Получаем текущие оси
ax.set_facecolor('#041804')  # Очень темный зеленый фон для осей

# Построение всех точек последовательности
plt.scatter(x, y, color='#207a20', label='Все значения последовательности', edgecolor='black', alpha=0.5)
plt.plot(x, y, color='#207a20', alpha=0.5)

# Построение выделенной сходящейся последовательности x_{2k}
plt.scatter(x_even, y_even, color='#00b050', label='Сходящаяся подпоследовательность $x_{2k}$', edgecolor='black', s=50, alpha=0.6)
plt.plot(x_even, y_even, color='#00b050', linestyle='--', alpha=0.7)

# Подсветка точек, попадающих в epsilon-окрестность
if n0_index is not None:
    x_within_epsilon = x_even[indices_within_epsilon]
    y_within_epsilon = y_even[indices_within_epsilon]
    plt.scatter(x_within_epsilon, y_within_epsilon, color='cyan', label='$x_{2k}$ в $\\varepsilon$-окрестности', edgecolor='white', s=70, alpha=0.5)

# Горизонтальная линия для предела
plt.axhline(y=limit, color='red', linestyle='--', linewidth=1.2, label='Предел подпоследовательности $\\frac{3\\pi}{2}$')

# Заголовки и метки осей
plt.title('Визуализация $\\varepsilon$-окрестности предела подпоследовательности $x_{2k}$ при $\\varepsilon = 0.1$', fontsize=14, color='#90ee90')
plt.xlabel('Значения $n \in \mathbb{N}$', fontsize=15, color='#90ee90')
plt.ylabel('Значения $x_{n}$', fontsize=12, color='#90ee90')

# Добавление сетки
plt.grid(visible=True, which='both', color='darkgreen', linestyle='--', linewidth=0.7, alpha=0.5)

# Показ значений на осях
plt.xticks(np.arange(num_X1, num_X2, 10), color='#90ee90')  # Шаг значений по оси x в светло-зеленом
plt.yticks(np.linspace(min(y) - 0.5, max(y) + 3, 7), color='#90ee90')  # Шаг значений по оси y в светло-зеленом

# Легенда с изменением цвета текста
legend = plt.legend(facecolor='#092b09')
for text in legend.get_texts():
    text.set_color('#90ee90')  # Установка цвета текста легенды на светло-зеленый

# Показ графика
plt.show()

# Вывод значения n0
if n0_index is not None:
    print(f'n0 = {x_even[n0_index]//2}, начиная с которого элементы подпоследовательности находятся в ε-окрестности предела.')
else:
    print('Не найдено значений в указанной ε-окрестности.')
```

n0 = 66 — первый индекс, начиная с которого элементы подпоследовательности попадают в ε-окрестность.
